# Annual Income Distribution

```python
from mpl_toolkits.mplot3d import Axes3D
plt.figure(figsize = (12,8))
sns.set(style = 'whitegrid')
sns.distplot(Mall['Annual Income (k$)'])
plt.title('Distribution of Annual Income (k$)', fontsize = 30)
plt.xlabel('Range of Annual Income (k$)')
plt.ylabel('Count')
```

# AGE DISTRIBUTION:-

```python
plt.figure(figsize=(12, 8))
sns.set(style = 'whitegrid')
sns.distplot(Mall['Age'])
plt.title('Distribution of Age', fontsize = 40)
plt.xlabel('Range of Age')
plt.ylabel('Count')
```

# SPENDING SCORE DISTRIBUTION

```python
plt.figure(figsize=(12, 6))
sns.set(style = 'darkgrid')
sns.distplot(Mall['Spending Score (1-100)'])
plt.title('Distribution of Spending Score (1-100)', fontsize = 40)
plt.xlabel('Range of Spending Score (1-100)')
plt.ylabel('Count')
```

```python
sns.distplot(Mall['Spending Score (1-100)'])
```

# GENDER ANALYSIS

```python
genders = Mall.Gender.value_counts()
sns.set_style("darkgrid")
plt.figure(figsize=(11,4))
sns.barplot(x=genders.index, y=genders.values)
plt.show()
```

```python
# Now i will show this in the Heatmap
```

```python
plt.rcParams['figure.figsize'] = (14, 8)
sns.heatmap(Mall[[ 'Gender', 'Age', 'Annual Income (k$)','Spending Score (1-100)']].corr(), cmap = 'magma_r', annot = True, linewidths=.5)
plt.title('Heatmap', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (16, 7)
g = sns.catplot(x="Gender", y="Spending Score (1-100)", kind="violin", inner=None, data=Mall)
sns.swarmplot(x="Gender", y="Spending Score (1-100)", color="k", size=3, data=Mall, ax=g.ax);
plt.title('Gender vs Spending Score', fontsize = 16)
plt.xlabel('Gender')
plt.ylabel('Spending Score (1-100)')
```
